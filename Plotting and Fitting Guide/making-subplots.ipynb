Subplots are particularly useful when not using jupyter notebooks as it will stop five plot windows from opening up.

```python
import matplotlib.pyplot as _plt
import numpy as _np
```

```python
experimentinput = _np.loadtxt("data.txt", delimiter=",")  # load input data from file, comma separated
experiment = _np.transpose(experimentinput)  # transpose so x, y, y-error are separate row vectors
x_data = experiment[0]  # for clarity, explicitly name vectors
y_data = experiment[1]
y_error = experiment[2]
```

```python
fig, (ax1, ax2) = _plt.subplots(1, 2, figsize=(12, 18))

ax1 = _plt.subplot2grid((3, 1), (1, 0), rowspan=1)
ax2 = _plt.subplot2grid((3, 1), (2, 0), rowspan=1)

fontsize = 20

ax1.errorbar(x_data, y_data, yerr=y_error, fmt='o', markersize=6)

ax1.set_title('OG Plot of Justice', fontsize=fontsize)
ax1.set_xlabel('x axis', fontsize=fontsize)
ax1.set_ylabel('y axis', fontsize=fontsize)

ax1.tick_params(axis='both', width=2, length=5, labelsize=18, direction='in')

_border_lines = [border_line.set_linewidth(2.) for border_line in ax1.spines.values()]

ax2.errorbar(x_data, y_data, yerr=y_error, fmt='o', markersize=6)

ax2.set_title('Plot of Justice 2', fontsize=fontsize)
ax2.set_xlabel('x axis', fontsize=fontsize)
ax2.set_ylabel('y axis', fontsize=fontsize)

ax2.tick_params(axis='both', width=2, length=5, labelsize=18, direction='in')

_border_lines = [border_line.set_linewidth(2.) for border_line in ax2.spines.values()]

fig.subplots_adjust(hspace=0.3)

_plt.show()
```

If you do not want to use subplots for whatever reason every attribute (labels, titles, linestyles) for a given figure will need to go after the plt.figure(x) (but before the next plt.figure) to be applied to that figure. To display all the plots once again use plt.show().

```python
# theory plot
x_values = _np.linspace(1,20,10)  # produce 10 equally spaced points between 1 and 20
y_values = x_values**2  # square each element in the vector
_plt.figure(1)  # First plot is figure1
_plt.plot(x_values, y_values)  # Plot theory data, line connects points
_plt.xlabel('Frogs')
_plt.ylabel('Unicorns')


_plt.figure(2)  # next plot is figure2
_plt.errorbar(x_data, y_data, yerr=y_error, fmt='kx')  # scatter plot with error bars, 'kx'=plot as black x
#plot(x_data, y_data,'k.')  # plots without error bars, 'k.' plots as black points
_plt.xlabel('Bears')
_plt.ylabel('Beets')
_plt.title('Battlestar Galatica')

# example of combining theory and data plot
_plt.figure(3)
_plt.plot(x_values, y_values)  # plot theory
_plt.errorbar(x_data, y_data, yerr=y_error, fmt='ro')  # on top, also plot experiment, 'ro'=plot as red o
_plt.xlabel("Kittens")  # Always label your plots!
_plt.ylabel("Pirates")
_plt.title("Pirates vs Kittens")
_plt.semilogx()  # now plot x on log axis
# xlim([0.5, 25])  # manual definition of x axis limits [xmin, xmax]
# ylim([-20, 450])  # manual definition of y axis limits [ymin, ymax]
_plt.show()
```
